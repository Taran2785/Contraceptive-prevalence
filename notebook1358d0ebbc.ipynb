Little MCAR test

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

women = pd.read_csv("/kaggle/input/contraceptive/contraceptive any method all women 15-49.csv")
married = pd.read_csv("/kaggle/input/contraceptive/contraceptive prevalence any method married women age 15-49.csv")
unmarried = pd.read_csv("/kaggle/input/contraceptive/contraceptive prevalence any method unmarried women age 15-49 .csv")

def little_mcar_test(df):
    """Perform Little's MCAR test manually."""
    
    # Select only numerical columns (MCAR test requires numerical data)
    df_numeric = df.select_dtypes(include=[np.number])
    
    # Drop rows where all values are missing
    df_numeric = df_numeric.dropna(how="all")
    
    # Mean imputation (replacing missing values with column means)
    df_filled = df_numeric.fillna(df_numeric.mean())
    
    # Compute chi-square test
    s, p_value = stats.chisquare(df_numeric.isnull().sum(), df_filled.isnull().sum())
    
    # Print results
    print(f"MCAR Test p-value: {p_value:.4f}")
    if p_value > 0.05:
        print("Fail to reject MCAR -> Missing data is MCAR (Completely at Random).")
    else:
        print("Reject MCAR -> Missing data is MAR or MNAR (Not Completely at Random).")

# Run the function for each dataset
little_mcar_test(women)
little_mcar_test(married)
little_mcar_test(unmarried)
```

Reshape Data (Convert Wide to Long Format)

To analyze MAR, we need to reshape the data from wide format to long format, where each row represents a Country-Year pair.

📌 Convert Wide to Long Format

```python
import pandas as pd
women= pd.read_csv("/kaggle/input/contraceptive/contraceptive any method all women 15-49.csv")
married= pd.read_csv('/kaggle/input/contraceptive/contraceptive prevalence any method married women age 15-49.csv')
unmarried = pd.read_csv("/kaggle/input/contraceptive/contraceptive prevalence any method unmarried women age 15-49 .csv")
def reshape_data_to_long(df, id_col='Country Name'):
    return df.melt(id_vars=[id_col], var_name="Year", value_name="Value")


# Example usage:
women_long = reshape_data_to_long(women)
# Ensure "Value" remains as it is (without forcing float conversion)
women_long["Value"] = pd.to_numeric(women_long["Value"], errors="coerce")  # Convert valid numbers, keep NaNs


married_long = reshape_data_to_long(married)
# Ensure "Value" remains as it is (without forcing float conversion)
married_long["Value"] = pd.to_numeric(married_long["Value"], errors="coerce")  # Convert valid numbers, keep NaNs

unmarried_long = reshape_data_to_long(unmarried)
# Ensure "Value" remains as it is (without forcing float conversion)
unmarried_long["Value"] = pd.to_numeric(unmarried_long["Value"], errors="coerce")  # Convert valid numbers, keep NaNs
```

Check If Missingness is Related to Country (MAR Test)

If missing values in years (e.g., 1990, 2005) are more frequent for certain countries, this suggests MAR.

📌 Check Missingness by Country

```python
def check_missing_by_country(df, country_col='Country Name', year_col='Year', value_col='Value'):
    missing_counts = df[df[value_col].isnull()].groupby(country_col).size()
    total_counts = df.groupby(country_col).size()
    
    missing_percentage = (missing_counts / total_counts) * 100
    missing_percentage = missing_percentage.dropna().sort_values(ascending=False)
    
    print(f"Percentage of Missing {value_col} by {country_col}:\n")
    print(missing_percentage)

# Example usage:
check_missing_by_country(women_long)
check_missing_by_country(married_long)
check_missing_by_country(unmarried_long)
```

Interpretation of Results:

Country Variable:

From the output, the percentage of missing values for different countries varies. Some countries have 100% missing values for the data points, while others have varying degrees of missing data, with Kenya and Philippines showing around 80-90% missing data.

100% Missing Data: Countries like Afghanistan, OECD members, Macao SAR, and Finland show 100% missing data, which means no data was reported for those countries in the dataset for the variable in question.

Missing Data for Other Countries: Countries like Kenya, Philippines, United Kingdom, and Peru have missing data percentages ranging from 70% to 80%.

MAR Test Interpretation for Country:

The countries with 100% missing data (e.g., Afghanistan, Finland, etc.) suggest that missing data might not be missing at random because the missingness could be attributed to certain reasons (e.g., lack of data reporting, inconsistent collection sources).

For the countries with lower percentages of missing data (Kenya, Philippines, etc.), the missingness might be linked to specific years (e.g., years of missing reports), and this could indicate Missing at Random (MAR).

The MAR hypothesis suggests that the probability of missingness depends on observed data, and it’s possible that these countries have missing data for specific years rather than having a random pattern.

Conclusion for Country:

Since there are countries with 100% missing data and others with varying degrees of missingness, it suggests that the data for these countries is likely not Missing Completely at Random (MCAR), but it might follow a Missing at Random (MAR) pattern depending on observed data (such as reporting inconsistencies across countries).

```python
def check_missing_by_year(df, year_col='Year', value_col='Value'):
    missing_counts = df[df[value_col].isnull()].groupby(year_col).size()
    total_counts = df.groupby(year_col).size()
    
    missing_percentage = (missing_counts / total_counts) * 100
    missing_percentage = missing_percentage.dropna().sort_values(ascending=False)
    
    print(f"Percentage of Missing {value_col} by {year_col}:\n")
    print(missing_percentage)

# Example usage:
check_missing_by_year(women_long)
check_missing_by_year(married_long)
check_missing_by_year(unmarried_long)
```

MAR Test Interpretation for Year:

The years with missing data percentages ranging from 80% to 90% (e.g., 2000, 2006, 2010) may be Missing at Random (MAR), where the missingness is likely dependent on other observed variables (like the country or the availability of reports for those years).

Conclusion for Year:

Similar to the country variable, the year data is likely not Missing Completely at Random (MCAR) because of the presence of 100% missing years.
However, for other years with a lower percentage of missing data, the data might be considered MAR, where the missingness could be explained by observed factors, like country or year-specific reporting gaps.

Overall Interpretation:
